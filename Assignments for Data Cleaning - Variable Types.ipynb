```python
import pandas as pd
import numpy as np
movie = pd.read_csv("https://djl-lms-assets.s3.eu-central-1.amazonaws.com/datasets/movie.csv")
movie.columns
movie_imdb=movie["imdb_score"]
#display(movie_imdb)
movie_imdb.describe()
movie_imdb.head
def make_discrete(imdb_score):
    if imdb_score >= 7.0:
        return 1
    elif (imdb_score < 7) and (imdb_score < 5):
        return 2
    else:
        return 3
movie['imdb_group'] = movie['imdb_score'].apply(make_discrete)
display(movie.head())
movie.groupby("imdb_group")["imdb_score"].count()
```
